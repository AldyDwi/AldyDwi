### 1. Import

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

### 2. Membaca dan menampilkan image

```python
# Membuat list untuk menyimpan url beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://unair.ac.id/wp-content/uploads/2020/12/Kucing.jpg"]

# Baca dan tampilkan image
# loop url image, disimpan pada list
for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('/n')
```

### 3. Melihat ukuran file image

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ", tinggi, "x", lebar)
cv2_imshow(image_2)
```

### 4. Memberikan garis horizontal di tengah image

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# buat garis horizontal
for y in range (lebar):
  image_3[int((tinggi)/2), y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

### Tugas

2. Buat garis vertikal

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# buat garis vertikal
for y in range (tinggi):
  image_3[y, int((lebar)/2)] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

2. Garis diagonal menyilang

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Garis diagonal 
for x in range(lebar):
    y = int((tinggi/lebar) * x)
    if y < tinggi:
        image_3[y, x] = [255,255,255]

# Garis diagonal 
for x in range(lebar):
    y = int(tinggi - 1 - (tinggi/lebar) * x)
    if 0 <= y < tinggi:
        image_3[y, x] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

3. Garis horizontal di tengah image

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# buat garis horizontal di tengah gambar
y_mid = tinggi // 2
start_x = 2 * lebar // 5 # mulai dari 2/5 lebar
end_x   = 4 * lebar // 5 # berhenti di 4/5 lebar

for x in range(start_x, end_x):
    image_3[y_mid, x] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

4. Kotak dalam gambar

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Tentukan ukuran sisi persegi (misal 1/2 dari sisi terpendek gambar)
sisi = min(tinggi, lebar) // 2

# Hitung titik kiri-atas & kanan-bawah supaya kotak di tengah
x1 = (lebar - sisi) // 2
y1 = (tinggi - sisi) // 2
x2 = x1 + sisi
y2 = y1 + sisi

# Gambar kotak putih
cv.rectangle(image_3, (x1, y1), (x2, y2), (255, 255, 255), -1)

final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```
